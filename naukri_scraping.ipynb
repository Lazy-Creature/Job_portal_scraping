```python
#import libraries
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service 
from selenium.webdriver.chrome.options import Options
```

```python
path=Service(r'C:\Users\Dinesh Halder\Downloads\chromedriver-win64\chromedriver.exe')
driver=webdriver.Chrome(service=path)
driver.maximize_window()
```

```python
# empty lists
job_title=[]
company_name=[]
rating =  []         
review =[]           
exp = []      
location=[]
salary=[]
description=[]
skill=[]
```

```python
for page in range(1,12):
    time.sleep(2)
    driver.get(f'https://www.naukri.com/data-science-jobs-{page}')
    time.sleep(3)
    all_jobs=driver.find_element(By.XPATH,"//div[@class='styles_job-listing-container__OCfZC']")
    single_job_frame=all_jobs.find_elements(By.XPATH,".//div[@class='cust-job-tuple layout-wrapper lay-2 sjw__tuple ']")
    for job in single_job_frame:
        try:
            job_title.append(job.find_element(By.XPATH,".//a[@class='title ']").text)
        except:
            job_title.append(None)
        
        
        try:
            company_name.append(job.find_element(By.XPATH,".//a[@class=' comp-name mw-25']").text)
        except:
            company_name.append(None)
              
                
        try:
            rating.append(job.find_element(By.XPATH,".//span[@class='main-2']").text)
        except:
            rating.append(None)    
        
        try:
            review.append(job.find_element(By.XPATH,".//a[@class='review ver-line']").text)
        except:
            review.append(None)
            
        try:
            exp.append(job.find_element(By.XPATH,".//span[@class='exp-wrap']").text)
        except:
            exp.append(None)
            
        try:
            location.append(job.find_element(By.XPATH,".//span[@class='ni-job-tuple-icon ni-job-tuple-icon-srp-location loc']").text)
        except:
            location.append(None)
        
        try:
            salary.append(job.find_element(By.XPATH,".//span[@class='ni-job-tuple-icon ni-job-tuple-icon-srp-rupee sal']").text)
        except:
            salary.append(None)
        
        try:
            description.append(job.find_element(By.XPATH,".//div[@class=' row4']").text)
        except:
            description.append(None)
        
        try:
            skill.append(job.find_element(By.XPATH,".//div[@class=' row5']").text.split())  
        except:
            skill.append(None)
            
            
driver.quit()   
```

```python
df=pd.DataFrame({'Title':job_title,'Company':company_name,'Rating':rating,'Review':review,'Experience':exp,'Location':location,'Salary':salary,'Description':description,'Skills':skill})
```
